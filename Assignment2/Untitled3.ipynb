```python
caption = widgets.Label(value='The slider value is in its initial position.')
slider = widgets.IntSlider(min=-5, max=5, value=1, description='Slider')

def handle_slider_change(change):
    caption.value = 'The slider value is ' + (
        'negative' if change.new < 0 else 'nonnegative'
    )

slider.observe(handle_slider_change, names='value')

display(caption, slider)
```

```python
import asyncio

class Timer:
    def __init__(self, timeout, callback):
        self._timeout = timeout
        self._callback = callback

    async def _job(self):
        await asyncio.sleep(self._timeout)
        self._callback()

    def start(self):
        self._task = asyncio.ensure_future(self._job())

    def cancel(self):
        self._task.cancel()

def debounce(wait):
    """ Decorator that will postpone a function's
        execution until after `wait` seconds
        have elapsed since the last time it was invoked. """
    def decorator(fn):
        timer = None
        def debounced(*args, **kwargs):
            nonlocal timer
            def call_it():
                fn(*args, **kwargs)
            if timer is not None:
                timer.cancel()
            timer = Timer(wait, call_it)
            timer.start()
        return debounced
    return decorator
```

```python
slider = widgets.IntSlider()
text = widgets.IntText()

@debounce(0.2)
def value_changed(change):
    text.value = change.new
slider.observe(value_changed, 'value')

widgets.VBox([slider, text])
```

```python
import asyncio
from time import time

def throttle(wait):
    """ Decorator that prevents a function from being called
        more than once every wait period. """
    def decorator(fn):
        time_of_last_call = 0
        scheduled, timer = False, None
        new_args, new_kwargs = None, None
        def throttled(*args, **kwargs):
            nonlocal new_args, new_kwargs, time_of_last_call, scheduled, timer
            def call_it():
                nonlocal new_args, new_kwargs, time_of_last_call, scheduled, timer
                time_of_last_call = time()
                fn(*new_args, **new_kwargs)
                scheduled = False
            time_since_last_call = time() - time_of_last_call
            new_args, new_kwargs = args, kwargs
            if not scheduled:
                scheduled = True
                new_wait = max(0, wait - time_since_last_call)
                timer = Timer(new_wait, call_it)
                timer.start()
        return throttled
    return decorator
```

```python
slider = widgets.IntSlider()
text = widgets.IntText()

@throttle(0.2)
def value_changed(change):
    text.value = change.new
slider.observe(value_changed, 'value')

widgets.VBox([slider, text])
```

```python
import vtk
from vtk.util.numpy_support import vtk_to_numpy
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# --- Load .vti file ---
reader = vtk.vtkXMLImageDataReader()
reader.SetFileName("Downloads/Assignment2_77d416ee-6f81-4276-a8ac-cb0f0a5c41b7/mixture.vti")  # Replace with your filename
reader.Update()
image_data = reader.GetOutput()

# --- Extract Data ---
dims = image_data.GetDimensions()
spacing = image_data.GetSpacing()
origin = image_data.GetOrigin()

array = image_data.GetPointData().GetArray("ImageFile")
data_np = vtk_to_numpy(array).reshape(dims[::-1])  # z, y, x

# --- Coordinate Grid ---
z, y, x = np.mgrid[0:data_np.shape[0], 0:data_np.shape[1], 0:data_np.shape[2]]
x = origin[0] + x * spacing[0]
y = origin[1] + y * spacing[1]
z = origin[2] + z * spacing[2]

# --- Define slider ---
min_val = float(np.min(data_np))
max_val = float(np.max(data_np))
default_val = float(np.percentile(data_np, 70))

isovalue_slider = widgets.FloatSlider(
    value=default_val,
    min=min_val,
    max=max_val,
    step=(max_val - min_val) / 100,
    description='Isovalue:',
    continuous_update=False,
    readout_format='.2f'
)

# --- Define update function ---
def update_isosurface(isovalue):
    fig = go.Figure(data=go.Isosurface(
        x=x.flatten(),
        y=y.flatten(),
        z=z.flatten(),
        value=data_np.flatten(),
        isomin=isovalue,
        isomax=isovalue,
        surface_count=1,
        caps=dict(x_show=False, y_show=False, z_show=False),
        colorscale='Viridis'
    ))
    fig.update_layout(
        title=f'Isosurface at Value {isovalue:.2f}',
        scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
        width=800,
        height=800
    )
    fig.show()

# --- Connect slider to function ---
widgets.interact(update_isosurface, isovalue=isovalue_slider)
```

```python
import vtk
from vtk.util.numpy_support import vtk_to_numpy
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# --- Load .vti file ---
reader = vtk.vtkXMLImageDataReader()
reader.SetFileName("Downloads/Assignment2_77d416ee-6f81-4276-a8ac-cb0f0a5c41b7/mixture.vti")  # Replace with your filename
reader.Update()
image_data = reader.GetOutput()

# --- Extract scalar data ---
dims = image_data.GetDimensions()
array = image_data.GetPointData().GetArray("ImageFile")
data_np = vtk_to_numpy(array).reshape(dims[::-1])  # (z, y, x)
flat_data = data_np.flatten()

# --- Get value range for the slider ---
min_val = float(flat_data.min())
max_val = float(flat_data.max())

# --- Define the slider widget ---
isovalue_slider = widgets.FloatSlider(
    value=np.percentile(flat_data, 70),
    min=min_val,
    max=max_val,
    step=(max_val - min_val) / 200,
    description='Isovalue:',
    continuous_update=True,
    readout_format='.2f',
    layout=widgets.Layout(width='600px')
)

# --- Function to update the histogram dynamically ---
def update_histogram(isovalue):
    # Define range around the isovalue
    lower = isovalue - 0.25
    upper = isovalue + 0.25

    # Filter data
    subset = flat_data[(flat_data >= lower) & (flat_data <= upper)]

    # Plot
    plt.figure(figsize=(10, 5))
    counts, bins, _ = plt.hist(
        subset, bins=30,
        color='blue', edgecolor=(0, 0, 0, 0)  # blue bars, transparent edges
    )
    
    # Custom X-ticks (based on data range)
    xticks = np.round(np.linspace(lower, upper, 7), 2)  # 7 ticks in the zoomed range
    plt.xticks(xticks)

    # Custom Y-ticks (0, 10k, 20k...)
    max_y = int(np.ceil(counts.max() / 10000.0) * 10000)
    yticks = np.arange(0, max_y + 1, 10000)
    ytick_labels = [f'{int(y/1000)}k' if y != 0 else '0' for y in yticks]
    plt.yticks(yticks, ytick_labels)

    # Labels & layout
    plt.title(f"Histogram of 'ImageFile' Values in Range [{lower:.2f}, {upper:.2f}]")
    plt.xlabel('Scalar Value')
    plt.ylabel('Voxel Count')
    plt.tight_layout()
    plt.show()

# --- Display interactive plot ---
widgets.interact(update_histogram, isovalue=isovalue_slider)
```
